# SQL

## 1. Apa saja nama station yang memiliki status `active`?

```python
# code here
df = client.query('''
SELECT COUNT(*)
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`
WHERE status = 'active'
''').to_dataframe()
df
```

Knowledge : Total active station is 78

## 2. Berapa banyak station yang berada di `council_district = 9` dan memiliki status `active`?

```python
# code here
df = client.query('''
SELECT COUNT(*)
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`
WHERE
  status = 'active'
  AND council_district = 9
''').to_dataframe()
df
```

Total active station at council district 9 is 47.

## 3. Bandingkan banyaknya station antara `power_type = solar` dengan `power type = non-metered`?

```python
# code here
df = client.query('''
SELECT power_type, COUNT(*)
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`
GROUP BY power_type
HAVING power_type = 'solar' OR power_type = 'non-metered'
''').to_dataframe()
df
```

Hasil query menyatakan power_type solar lebih banyak yaitu 79 sementara non-metered berjumlah 3.

# Pandas

## 1. Tampilkan summary data! Berikan insight terkait summary data yang ditemukan!

```python
df = client.query('''
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`
''').to_dataframe()
df
```

Total data pada dataset ada 102 dengan data column yang akan digunakan untuk menjawab soal ini adalah number_of_docks yang merupakan integer.

```python
df.dropna(subset=['number_of_docks'], inplace=True)
```

## 2. Buatkan kolom baru `keterangan` yang memiliki ketentuan berikut

- Jika `number_of_docks` antara 0 - 5 -> value `station kecil`
- Jika `number_of_docks` antara 6 - 10 -> value `station sedang`
- Jika `number_of_docks` lebih dari/sama dengan 11 -> value `station besar`

```python
def kategoriStation(n):
  if n >= 0 and n <= 5 :
    return 'kecil'
  elif n >= 6 and n <= 10:
    return 'sedang'
  elif n >= 11:
    return 'besar'
  return '-'

keterangan = []
for data in df['number_of_docks'] :
  result = kategoriStation(data)
  keterangan.append(result)

df['keterangan'] = keterangan
df
```

Insight:
Dari dataset terdapat data null atau kosong sehingga harus dibersihkan dahulu.
Total data yang diproses dengan fungsi kategoriStation adalah 82.

## 3. Tampilkan nama station yang memiliki `power_type = solar` dan `council_district = 10`

```python
data3 = df[(df['power_type'] == 'solar') & (df['council_district'] == 10)]
data3['name']
```

Insight: Ditemukan 2 data yaitu Lake Austin Blvd @ Deep Eddy dan Lake Austin & Enfield sebagai station dengan power type solar yang terletak di council district 10.
